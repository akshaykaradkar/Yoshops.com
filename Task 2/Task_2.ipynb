```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
color = sns.color_palette("pastel")

import warnings
# current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')

import missingno as msno # missing data visualization module for Python
import pandas_profiling

#loading dataset
df = pd.read_csv('review_dataset.csv')
df
```

```python
#Check for missing values
msno.bar(df)

# Get the count of missing values in each column
missing_values_count = df.isnull().sum().sort_values(ascending=False)

# Print the columns with missing values and the count of missing values
print(missing_values_count[missing_values_count > 0])

# Extract the rating from the 'stars' column
df['rating'] = df['stars'].str.extract(r'(\d+.\d+)').astype('float')

# Create a bar plot of the number of reviews given by customers for each rating
plt.figure(figsize=(15,6))
sns.countplot(df['rating'])
plt.xlabel('Rating')
plt.ylabel('Reviews Count')
plt.title('Reviews given by Customers')
plt.savefig("Reviews given by Customers.pdf")
plt.show()

# Create a pandas_profiling report for the data
profile_new = df.profile_report(title='Review Data Report')
profile_new.to_file("Review_Data_Report.html")
```

```python
# Load the data
df1 = pd.read_csv('orders_2016-2020_Dataset.csv')
df1
```

```python
# Rename the columns for better readability
df1.rename(index=str, columns={'Order #': 'order_num',
                              'Tracking #' : 'tracking_num',
                              'Order Date and Time Stamp' : 'timespan'}, inplace=True)

# Check for missing values
missing_values = df1.isnull().sum().sort_values(ascending=False)
print(missing_values)

# Drop rows with missing values
df1.dropna(subset=['Payment Method'], inplace=True)

# Create a new dataframe with relevant columns
payment_df = df1[['order_num', 'Payment Status', 'Payment Method']]

# extract the mode of payment and amount from the 'Payment Method' column
payment_df['Mode of Payment'] = payment_df['Payment Method'].str.split(pat = " ", expand=True)[0]
payment_df['Amount'] = payment_df['Payment Method'].str.split(pat = " ", expand=True)[1]

# Convert 'timespan' column to datetime object
df1['timespan'] = pd.to_datetime(df1['timespan'])

sns.set_palette('pastel')

# Create a bar chart to visualize the different payment methods
payment_count = payment_df['Mode of Payment'].value_counts()
plt.figure(figsize=(15,6))
sns.barplot(x=payment_count.index, y=payment_count.values)
plt.xlabel('Payment Method')
plt.ylabel('Number of Orders')
plt.title('Different Types of Payment Methods used in {}'.format(df1['timespan'].dt.year.iloc[0]))
plt.savefig("Different Types of Payment Methods in {}.pdf".format(df1['timespan'].dt.year.iloc[0]))
plt.show()


# Create a pandas_profiling report for the data
profile = df1.profile_report(title='Orders Data Profile Report')
profile.to_file("Orders_Data_Profile_Report.html")
```

```python
#analysis of Top Consumer States of India

states = df1[['order_num','Billing City', 'Billing State', 'Billing Zip']]
states.head()

states.isnull().sum().sort_values(ascending=False)
states.dropna(inplace= True)
states.isnull().sum().sort_values(ascending=False)
states['Billing State'].value_counts()
top_consumer_states = states.groupby('Billing State').count().sort_values("order_num", ascending=False).reset_index()
top_states = top_consumer_states[['order_num','Billing State']]
top_states.head()

sns.set_palette('pastel')

plt.figure(figsize=(15,6))
sns.barplot(data = top_states , x = "Billing State", y = "order_num")
plt.xlabel('Name of Consumer States')
plt.ylabel('Number of Orders')
plt.title('Top Consumer States of India')
plt.savefig("Top Consumer States of India.pdf")
plt.show()
```

```python
# Analysis of top consumer cities of India
cities = df1[['order_num', 'Billing City', 'Billing State']]
cities.head()

# Lowercase the city names for consistency
cities['Billing City'] = cities['Billing City'].str.lower()

# Group the data by city and state and count the number of orders
top_consumer_cities = cities.groupby(['Billing City', 'Billing State']).count().sort_values("order_num", ascending=False).reset_index()

# Select only the top 10 cities
top_cities = top_consumer_cities.head(10)

# Set the style for the plot
sns.set_style("whitegrid")

# Create the plot
plt.figure(figsize=(15,6))
sns.barplot(data = top_cities, x = "Billing City", y = "order_num", hue = "Billing State", dodge = False)
plt.xlabel('Name of Consumer Cities', fontsize = 14)
plt.ylabel('Number of Orders', fontsize = 14)
plt.title('Top Consumer Cities of India', fontsize = 18)
plt.legend(title = "State", fontsize = 14)
plt.xticks(rotation = 90)
plt.savefig("Top Consumer Cities of India.pdf")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter out rows where category is not relevant to the analysis
df_cat = df[df['category'].isin(['Mobiles', 'Toys & Games', 'Headphones', 'Accessories', 'Fashion', 'Personal Care', 'Laptops', 'Mobile Accessories', 'Womens'])]

# Check for missing values
print(df_cat.isnull().sum().sort_values(ascending=False))

# Clean up data and create a new column for rating
df['rating'] = df['stars'].str.extract(r'(\d\.\d)')

# Group by category and count the number of reviews
category = df_cat.groupby('category').count().sort_values("status", ascending=False).reset_index()
category_new = category[['category','status']]

# Plot a bar chart using Seaborn
plt.figure(figsize=(15,6))
sns.barplot(data = category_new.head(10) , x = "category", y = "status", palette='pastel')
plt.xlabel('Name of the Category')
plt.ylabel('Number of Reviewed Status')
plt.title('Top Selling Product Category')
plt.savefig("Top Selling Product Category.pdf")
plt.show()
```

```python
# Check the columns in the dataframe
print(df.columns)

# Group by category and count the number of ratings
prod_category = df.groupby('category').count().sort_values("rating", ascending=False).reset_index()
category_rating = prod_category[['category','rating']]

# Plot a bar chart using Seaborn
plt.figure(figsize =(15,10))
sns.barplot(data=category_rating, x="rating", y="category",palette='pastel')
plt.xlabel('Total Rating')
plt.ylabel('Name of Categories')
plt.title('Analysis of Reviews for All Product Categories')
plt.savefig("Top rated Product Category.pdf")
plt.show()
```

```python
#analysis of Number of Orders Per Month Per Year
df1.head()
time_data = df1[['order_num','timespan']]
time_data.head()
#converting into time stamp
time_data['timespan'] = pd.to_datetime(time_data['timespan'])

print(time_data['timespan'].dtype)

# Extract the year, month, day, hour, day_name, and month_name from the timespan column
time_data['year'] = pd.to_datetime(time_data['timespan']).dt.year
time_data['month'] = pd.to_datetime(time_data['timespan']).dt.month
time_data['day'] = pd.to_datetime(time_data['timespan']).dt.day
time_data['hour'] = pd.to_datetime(time_data['timespan']).dt.hour
time_data['day_name'] = pd.to_datetime(time_data['timespan']).dt.day_name()
time_data['month_name'] = pd.to_datetime(time_data['timespan']).dt.month_name()

year_data = time_data.groupby('year').count().sort_values("order_num", ascending=False).reset_index()
year_data.head()

orders_per_year = year_data[['year','order_num']]

orders_per_year.head()

plt.figure(figsize=(15,6))
sns.barplot(data = orders_per_year , x = "year", y = "order_num")
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('Yearly Orders Analysis')
plt.savefig("Number of Orders per Year.pdf")
plt.show()
```

```python
# Group by year and month and count the number of orders
month_year_data = time_data.groupby(['year', 'month_name']).count().reset_index()

# Plot a heatmap using seaborn
plt.figure(figsize=(15,10))
sns.heatmap(month_year_data.pivot('month_name','year','order_num'), annot=True, fmt="f",cmap='YlGnBu')
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Number of Orders Per Month Per Year')
plt.savefig("Number of Orders Per Month Per Year.pdf")
plt.show()
```

```python
# Group by month_name and count the number of orders
month_data = time_data.groupby('month_name').count().sort_values("order_num", ascending=False).reset_index()
orders_per_month = month_data[['month_name','order_num']]

# Plot a bar chart using Seaborn
plt.figure(figsize=(15,6))
sns.barplot(data = orders_per_month , x = "month_name", y = "order_num", palette='pastel')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Monthly Order Analysis')
plt.savefig("Number of Orders per Month.pdf")
plt.show()
```

```python
# group by year and month and count number of orders
per_month_per_year_data = time_data.groupby(['year' , 'month_name']).count().sort_values("year", ascending=False).reset_index()
per_month_per_year = per_month_per_year_data[['year','month_name','order_num']]

# plot a bar chart using Seaborn
plt.figure(figsize=(15,6))
sns.barplot(data = per_month_per_year , x = "year", y = "order_num", hue="month_name",palette='pastel')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Per Month Per Year Analysis')
plt.savefig("Number of Orders per Month per Year 2.pdf")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data = per_month_per_year , x = "month_name", y = "order_num", hue="year")
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Per Month Per Year Analysis')
plt.savefig("Number of Orders per Month per Year 3.pdf")
plt.show()
```

```python
# Plot a 3D bar chart using Seaborn
from mpl_toolkits.mplot3d import Axes3D

# Group by year, month and count the number of orders
per_month_per_year_data_2 = time_data.groupby(['year' , 'month']).count().sort_values("year", ascending=False).reset_index()
per_month_per_year_2 = per_month_per_year_data_2[['year','month','order_num']]

# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D bar chart
ax.bar3d(per_month_per_year_2['year'], per_month_per_year_2['month'], [0]*len(per_month_per_year_2), 1, 1, per_month_per_year_2['order_num'])

# Set the labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Month')
ax.set_zlabel('Number of Orders')
plt.title('Number of Orders Per Month Per Year Analysis')
plt.savefig("Number of Orders per Month per Year 4.pdf")
plt.show()
```

```python
#analysis of Number of Orders Across Parts of a Day
orders_per_day = time_data[['day_name','hour','order_num']]
orders_per_day.head()
orders_per_day = time_data.groupby(['hour']).count().sort_values("order_num", ascending=False).reset_index()
orders_per_day_data = orders_per_day[['hour' ,'order_num']]
orders_per_day_data.head(25)
plt.figure(figsize =(15,6))
sns.barplot(data=orders_per_day_data , x = "hour" , y = "order_num" )
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Across Parts of a Day')
plt.savefig("Hourly Analysis.pdf")
plt.show()
```

```python
#dataset for per hour per day
order_timestamps =df1[['order_num','timespan','Fulfillment Status']]
order_timestamps['timespan'] = pd.to_datetime(order_timestamps['timespan'])
order_timestamps['hour'] = pd.to_datetime(order_timestamps['timespan']).dt.hour
order_timestamps['day_name'] = pd.to_datetime(order_timestamps['timespan']).dt.day_name()

#analysis of Number of Orders Across Parts of a Day
orders_per_day = order_timestamps.groupby(['day_name','hour']).count().sort_values("order_num", ascending=False).reset_index()
orders_per_day_data = orders_per_day[['day_name','hour' ,'order_num']]

plt.figure(figsize =(15,6))
sns.lineplot(data=orders_per_day_data , x = "hour" , y = "order_num", hue="day_name",  marker='o')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Across Parts of a Day')
plt.savefig("Hourly Analysis 2.pdf")
plt.show()
```
