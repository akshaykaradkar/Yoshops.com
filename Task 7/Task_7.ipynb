```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

baseurl = "https://www.careers360.com/"

headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.59'
    }

allcollegedata = []
collegelinks = []
collegelist = []

for x in range(0,532):
    print(f"Scraping data from page no. {x} ")
    r = requests.get(f'https://www.careers360.com/colleges/india-colleges-fctp?sort=popularity&page={x}')
    
    soup = BeautifulSoup(r.text,'lxml')
    
    collegelist = soup.find_all('div' , class_ ='content')
    
    for college in collegelist:
        name = college.h2.a.text
        state = college.ul.li.get_text()
        details = college.find('ul', attrs = {'class' : 'rank'}).text
        specl = college.find('div', attrs = {'class' : 'course'}).a.text.strip()
        website = college.h2.find('a')['href']
            
        collegedata = {
            'College_Name' : name,
            'College_State_City' : state,
            'College_Specialisation' : specl,
            'Website' : website,
            'College_Details' : details
        }
            
        allcollegedata.append(collegedata)
            
        print('Saving: ', collegedata['College_Name'])

college_df = pd.DataFrame(allcollegedata)
```

```python
college_df
```

```python
college_df2 = college_df.copy()
```

```python
def extract_city_state(a):
    asplit = a.split(",")
    city = asplit[0].split()[-1]
    state = asplit[1].split()[0]
    return city, state

college_df3 = college_df2.join(
    college_df2['College_State_City'].apply(
        lambda x: pd.Series(extract_city_state(x), index=["City", "State"])
    )
)
```

```python
def extract_ownership(a):
    asplit = a.split("\n")
    ownership = asplit[2].split()[0]
    reviews = asplit[3].split()[0]
    return ownership, reviews

college_df4 = college_df3.join(
    college_df3['College_Details'].apply(
        lambda x: pd.Series(extract_ownership(x), index=["Ownership", "No Of Reviews"])
    )
)
```

```python
del college_df4['College_State_City']
del college_df4['College_Details']
```
